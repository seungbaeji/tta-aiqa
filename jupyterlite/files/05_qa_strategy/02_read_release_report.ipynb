## 기본 개념: release gate와 재평가 조건

Release gate는 후보 모델이나 배포 변경이 운영으로 넘어가도 되는지 확인하는 기준입니다. MLOps에서 gate는 사람의 감으로 결정하는 절차가 아니라, metric threshold, live deployment check, observability readiness, rollback condition을 함께 보는 운영 규칙입니다.

보류는 실패와 다릅니다. 보류는 어떤 기준이 미달했거나 어떤 핵심 근거가 미확인이라서, owner와 next action을 정하고 같은 규칙으로 다시 평가하겠다는 상태입니다.

Release report는 의사결정의 감사 추적입니다. 누가 봐도 어떤 근거로 승인 또는 보류 의견이 나왔는지, 어떤 리스크가 남았는지, 어떤 조건에서 다시 판단할지 확인할 수 있어야 합니다.

| Report 항목 | 의미 | 확인해야 하는 이유 |
| --- | --- | --- |
| failed checks | 기준을 통과하지 못한 항목 | 보류 사유를 숫자로 설명 |
| unresolved risks | 아직 확인하지 못한 리스크 | 과장된 승인 방지 |
| owner action | 누가 무엇을 확인할지 | 후속 조치 추적 |
| rollback condition | 되돌림 기준 | 배포 후 장애 대응 |
| reevaluation rule | 다시 볼 기준 | 임의 기준 완화 방지 |
